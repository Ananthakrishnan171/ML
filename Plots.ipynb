**SCATTER PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.scatterplot(x="total_bill",y = "tip",data=tips)
plt.title("Scatter Plot of Total vs Tips")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

**VIOLIN PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.violinplot(x="total_bill",y = "tip",data=tips)
plt.title("Violin Plot of Total vs Tips")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

**BAR PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.barplot(x="total_bill",y = "tip",data=tips)
plt.title("Bar Plot of Total vs Tips")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

**HEATMAP PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
n_iris = iris.select_dtypes(include=['number'])
correlation_matrix = n_iris.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```

**PAIR PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
sns.pairplot(iris)
plt.title('pair plot of Iris Dataset')
plt.show()
```

**JOINT PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip', data=tips,kind = "hex")
plt.title('joint plot of Iris Dataset')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tips ($)')
plt.show()
```
